### 01) WAP to demonstrate the use of Pie chart.

```python
y = [35, 25, 25, 15]
label_y=["x","y","z","a"]
color=["r","y","b","g"]
ex=[0.2,0,0,0]
plt.pie(y,labels=label_y,colors=color,explode=ex,shadow=True)
plt.title("pie chart")
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x=["Math","OOP","Python","C"]
y=[90,98,99,85]
plt.bar(x,y,width=0.5)
for i in range(len(y)):
    plt.text(x[i],y[i],y[i])
# plt.barh(x,y)
# plt.grid()
plt.show()
```
